```python
import pandas as pd
import numpy as np
```

```python
header_list =  ['datetime','road_number','km','direction','all_units','inflow_units','outflow_unit','samecell_units','avg_speed','max_speed','avg_traveltime','max_traveltime']
```

```python
import glob
# get data file names

month = '05' ## select month
date = '01' ##select date

path =r'/content/drive/MyDrive/ Project/2020/2020'+'-'+month
filenames = sorted(glob.glob(path + '/data_2020' + '-'+ month + '-' +'?????.csv'))

day1 = 31
day2 = 31

day1 = 24*day1 - 24
day2 = 24*day2

dfs=[]

'''for filename in range(10): 
    dfs.append(pd.read_csv(filenames[filename],names=header_list))'''

for filename in range(day1,day2):
  print('loading ',filenames[filename])
  dfs.append(pd.read_csv(filenames[filename],names=header_list))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)

print('-----------completed---------------')
day1 = (day1+24)/24
day2 = day2/24
```
