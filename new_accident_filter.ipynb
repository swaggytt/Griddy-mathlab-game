------------------------------------------------------------- Accident zone

```python
df = pd.read_csv('/content/drive/MyDrive/ Project/NewAccidentData/tbl_rtddi.csv')
```

```python
df.dtypes
```

```python
df['DateRec'] = pd.to_datetime(df['DateRec'])
```

```python
df = df[df['DateRec']>='2020-1-1']
```

```python
df.drop(['id', 'DEAD_YEAR(Budha)','DEAD_YEAR','Age','Sex','BirthYear','NationalityId','Tumbol','District','Province','RiskHelmet','RiskSafetyBelt','DeadDate','ICD-10','created_at','verify_date','created_by'], axis=1, inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
df['lat'] = 0
df['lon'] = 0
df['road'] = 0
```

```python
df['lat'] = df['AccLat'].round(3)
df['lon'] = df['AccLong'].round(3)
df["coor"] = df.apply(lambda row: f"{row['lat']:.3f}, {row['lon']:.3f}", axis='columns' )
```

```python
df
```

```python
fig = px.scatter_mapbox(df, lat="AccLat", lon="AccLong")
fig.update_layout(mapbox_style="open-street-map") 
fig.show()
```
